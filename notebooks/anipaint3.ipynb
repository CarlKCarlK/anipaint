```python
import logging
from pathlib import Path
import math

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from PIL.ImageDraw import Draw

from anipaint import paint

folder = Path(r"E:\Dropbox\Watercolor Animation Assets")
output_folder = folder / "SkinMatte/Comp 2/outputs/run5_test"
matte_path = folder / "SkinMatte/Comp 2/Comp 2_00000.jpg"
brush_pattern = folder / "brushes/*.png"

logging.basicConfig(level=logging.INFO)
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
matte_image = Image.open(matte_path)
plt.imshow(matte_image)
plt.show()
```

```python
im_all = paint(
    matte_path,
    brush_pattern,
    random_count=200,
    keep_threshold=0,
    candidate_range=(1,256),
    credit_range=(1, 256),
    mixing_range=(255, 256),
    sprite_factor_range=(.5,1)
)
plt.imshow(im_all)
plt.show()
im0 = matte_image.copy()
im0.paste(im_all, (0, 0), im_all)
plt.imshow(im0)
plt.show()
```

```python
im_edge = paint(
    matte_path,
    brush_pattern,
    random_count=500,
    keep_threshold=.1,
    candidate_range=(10, 11),
    credit_range=(1, 30),
    mixing_range=(255, 256),
    sprite_factor_range=(.1,.5)
)
print("done")
```

```python
plt.imshow(im_edge)
plt.show()
im1 = matte_image.copy()
im1.paste(im_edge, (0, 0), im_edge)
plt.imshow(im1)
plt.show()
```

```python
im_edge2 = paint(
    matte_path,
    brush_pattern,
    random_count=200,
    keep_threshold=.25,
    candidate_range=(20, 21),
    credit_range=(10, 30),
    mixing_range=(255, 256),
    sprite_factor_range=(.1,1.5),
)

plt.imshow(im_edge2)
plt.show()

im2 = matte_image.copy()
im2.paste(im_edge2, (0, 0), im_edge2)
im2.paste(im_edge, (0, 0), im_edge)
plt.imshow(im2)
plt.show()
```

```python
im_in = paint(
    matte_path,
    brush_pattern,
    random_count=200,
    keep_threshold=.25,
    candidate_range=(21, 256),
    credit_range=(21, 256),
    mixing_range=(21, 256),
    sprite_factor_range=(.1,2),
)

plt.imshow(im_in)
plt.show()
im3 = matte_image.copy()
im3.paste(im_in, (0, 0), im_in)
im3.paste(im_edge2, (0, 0), im_edge2)
im3.paste(im_edge, (0, 0), im_edge)
plt.imshow(im3)
plt.show()
```
